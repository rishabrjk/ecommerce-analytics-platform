```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

# 📊 E-Commerce Analytics Project
### Author: Rishab Kothari

This project analyzes transaction data to extract business insights such as revenue trends and customer segmentation.


## 📁 Dataset Overview

This dataset contains transactional data including:
- Total bill (simulating revenue)
- Customer attributes
- Day and time of purchase

We use this to simulate e-commerce analytics.

```python
df['Revenue'] = df['total_bill']

revenue_by_day = df.groupby('day')['Revenue'].sum()
revenue_by_day
```

```python
revenue_by_day.plot(kind='bar')
plt.title("Revenue by Day")
plt.xlabel("Day")
plt.ylabel("Revenue")
plt.show()
```

```python
df['Segment'] = pd.qcut(df['Revenue'], 3, labels=["Low", "Medium", "High"])

df['Segment'].value_counts()
```

```python
df['Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Segments")
plt.show()
```

## 📈 Key Insights

- Revenue varies across different days, indicating peak business periods
- Customers can be segmented into low, medium, and high-value groups
- High-value customers contribute a significant portion of total revenue

These insights can help businesses improve targeting and decision-making.
